```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/household_dataset_500.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
mx=df["Household_Income"].max()
nx=df['Household_Income'].min()
range=mx-nx
range
```

```python
q3=df["Household_Income"].quantile(0.75)
q1=df["Household_Income"].quantile(0.25)
iqr=q3-q1
iqr
```

```python
sns.histplot(data=df,x="Household_Income",kde=True)
```

```python
from scipy.stats import norm
```

```python
mu, std = norm.fit(df['Household_Income'])
plt.figure()
sns.histplot(df['Household_Income'], kde=False, stat="density")
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p)
plt.title(f"Normal Fit: mu={mu:.2f}, std={std:.2f}")
plt.show()
```

```python
print("Skewness:", df['Household_Income'].skew())
print("Kurtosis:", df['Household_Income'].kurt())
```

```python
plt.figure()
sns.boxplot(data=df,x='Education_Level', y='Household_Income')
plt.title("Income by Education Level")
plt.show()
```

```python
plt.figure()
sns.boxplot(data=df,x='Urban_Rural', y='Household_Income')
plt.title("Income by Urban/Rural Area")
plt.show()
```

```python
plt.figure()
sns.histplot(data=df, x='Household_Income',hue='Education_Level' , kde=True)
plt.title("Income Distribution by Education")
plt.show()
```
